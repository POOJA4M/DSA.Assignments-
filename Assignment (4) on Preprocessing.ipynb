#### 4. Fill in all the missing values present in all the columns in the dataset.

#### 5. Check and handle outliers in atleast 3 columns in the dataset

```python
#Definition for checking outliers:
def col_outlier (df,col,threshold):
    z_score = (col-col.mean())/(col.std())
    column_outlier = df[abs(z_score)>threshold]
    return column_outlier
```

```python
col_outlier(data,data.Survived,2.5)
```

```python
col_outlier(data,data.Age,2.5)
```

```python
#Definition for handling outliers:
def cleaned_col (df,col,threshold):
    z_score = (col-col.mean())/(col.std())
    column_clean = df[abs(z_score)<=threshold]
    return column_clean
```

```python
cleaned1 = cleaned_col(data,data.Age,2.5)
cleaned1
```

```python
col_outlier(data,data.Fare,2.5)
```

```python
cleaned2 = cleaned_col(cleaned1,data.Fare,2.5)
cleaned2
```

```python
cleaned_data = cleaned2
```

#### 6. Do min max scaling on the feature set(Take 'Survived' as target)

```python
##Encoding
```

```python
name = pd.DataFrame(cleaned_data.Name)
sex = pd.DataFrame(cleaned_data.Sex)
ticket = pd.DataFrame(cleaned_data.Ticket)
cabin = pd.DataFrame(cleaned_data.Cabin)
embarked = pd.DataFrame(cleaned_data.Embarked)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

en_name = label_encoder.fit_transform(name)
en_sex = label_encoder.fit_transform(sex)
en_ticket = label_encoder.fit_transform(ticket)
en_cabin = label_encoder.fit_transform(cabin)
en_embarked = label_encoder.fit_transform(embarked)
cleaned_data.Name = en_name
cleaned_data.Sex = en_sex
cleaned_data.Ticket = en_ticket
cleaned_data.Cabin = en_cabin
cleaned_data.Embarked = en_embarked
cleaned_data
```
